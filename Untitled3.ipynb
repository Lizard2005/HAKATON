```python
def split_string(text,separator = ', '):
    # Разделение строки по символам ; и :
    parts = re.split('; |:', text)
    # Удаление пустых строк из результата, если они есть
    parts = [part.strip() for part in parts if part.strip()]
    return separator.join(parts[::2])
```
