```python
def find_decimal_numbers(text):
    # Регулярное выражение для поиска чисел с одной точкой
    pattern = r'\b\d+\.\d+\b'
    # Использование re.findall для поиска всех совпадений в строке
    decimal_numbers = re.findall(pattern, text)
    # Преобразование всех найденных строковых значений в числа с плавающей точкой
    return [float(number) for number in decimal_numbers]
```

```python
print(find_decimal_numbers('76.29.50: Инфекционные болезни; 34.41.02: Общие закономерности морфогенеза, эмбриогенеза и онтогенеза человека и животных'))
```
